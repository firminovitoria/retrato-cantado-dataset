```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "data/frases-musicas-final.csv"
```

```python
df = pd.read_csv(file_path)
df
```

```python
df["tem_padrao"].fillna("N", inplace=True)
df
```

```python
df.to_csv(file_path.replace(".csv", "-filled.csv"), index=False)
```

```python
df["tem_padrao"].value_counts()
```

```python
df["tem_padrao"].value_counts(normalize=True)
```

```python
ax = df['tem_padrao'].value_counts().plot(kind='bar', color=['red', 'blue'])
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
```
